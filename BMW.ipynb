```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import pycountry
import kagglehub
import random
```

```python
file_path = "/content/BMW sales data (2010-2024) (1).csv"
df = pd.read_csv(file_path)
print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.describe(include='all').T
```

```python
df = df.dropna(how='all')
df.columns = [c.strip().replace(" ", "_") for c in df.columns]
```

```python
if "Year" in df.columns:
    df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
```

```python
if not "Country" in df.columns:
    countries = [
        "United States", "Germany", "United Kingdom", "India",
        "China", "Japan", "France", "Canada", "Australia", "Brazil"
    ]
    df["Country"] = [random.choice(countries) for _ in range(len(df))]
else:
    print("'Country' column already exists:", df["Country"].unique()[:10])
```

```python
print(" Dataset Loaded Successfully!")
print("Shape:", df.shape)
df.head()
```

```python
fig = px.line(
    df.groupby("Year")["Sales_Volume"].sum().reset_index(),
    x="Year", y="Sales_Volume",
    title="📈 BMW Global Sales Over Time (2010–2024)",
    markers=True,
    color_discrete_sequence=["#1f77b4"]
)
fig.show()
```

```python
if "Model" in df.columns:
    model_sales = df.groupby("Model")["Sales_Volume"].sum().reset_index().sort_values(by="Sales_Volume", ascending=False).head(10)
    fig = px.bar(
        model_sales, x="Model", y="Sales_Volume",
        title="🏆 Top 10 Best-Selling BMW Models (2010–2024)",
        color="Sales_Volume", color_continuous_scale="Blues"
    )
    fig.show()
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).corr()
fig = px.imshow(
    numeric_cols, text_auto=True, color_continuous_scale="Blues",
    title="🔍 Correlation Matrix of Numeric Features"
)
fig.show()
```

```python
if "Price_USD" in df.columns:
    avg_price_country = df.groupby(["Year", "Country"])["Price_USD"].mean().reset_index()
    fig = px.line(
        avg_price_country, x="Year", y="Price_USD", color="Country",
        title="📊 Average BMW Price Trend by Country (2010–2024)"
    )
    fig.show()
```

```python
# COUNTRY-WISE SALES
country_sales = df.groupby("Country", as_index=False)["Sales_Volume"].sum()

def get_iso(country):
    try:
        return pycountry.countries.lookup(country).alpha_3
    except:
        return None

country_sales["iso_alpha"] = country_sales["Country"].apply(get_iso)
country_sales = country_sales.dropna(subset=["iso_alpha"])
fig = px.choropleth(
    country_sales,
    locations="iso_alpha",
    color="Sales_Volume",
    hover_name="Country",
    color_continuous_scale="Blues",
    title="🌍 BMW Sales Distribution by Country (2010–2024)"
)
fig.show()
```

```python
fig = px.pie(
    country_sales,
    names="Country",
    values="Sales_Volume",
    title="🧩 BMW Market Share by Country (2010–2024)"
)
fig.show()
```
